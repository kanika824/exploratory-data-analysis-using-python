```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observation: Zomato maximum records or transaction are from india and after that from usa and then from united kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

##OBSERVATION


1.   When rating is b/w 4.5 to 4.9--->excellent
2.    When rating is b/w 4.0 to 3.4--->excellent


1.    When rating is b/w 3.5 to 3.9--->excellent
2.    When rating is b/w 3.0 to 3.4--->excellent

1.   When rating is b/w 2.5 to 2.9--->excellent
2.    When rating is b/w 2.0to 2.4--->excellent







```python
ratings.head()
```

```python
 import matplotlib
 matplotlib.rcParams['figure.figsize']=(12,6)
 sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,color='blue')
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings ,palette=['blue','red','orange','yellow','green','green'])
```

# Observations:


1. not rated count is very high

```python
## count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
